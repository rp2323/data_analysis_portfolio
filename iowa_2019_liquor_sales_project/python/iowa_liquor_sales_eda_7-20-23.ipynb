```python
updated_df['Date'] = pd.to_datetime(updated_df['Date'])
# ^ change 'Date' column data type from 'object' to 'datetime'
updated_df['day_of_week'] = updated_df['Date'].dt.dayofweek
# ^ set new column 'day_of_week' equal to the dayofweek version of the date
sales_by_dow = updated_df.groupby('day_of_week')['Invoice/Item Number'].count()
# ^ set sales_by_dow variable equal to the updated_df table, grouping by day of week with another column set as counts of invoices
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# ^ updated day of week indexes so they're the literal day names rather than numbers
sales_by_dow.index = days
sales_by_dow
```

```python
grouped_data = updated_df.groupby('Category Name')['Sale (Dollars)'].sum()
# ^ Group data by category and sum
sorted_gd_cat = grouped_data.sort_values(ascending=False)
# ^ display grouped data in descending order to identify top category
sorted_gd_cat
```
